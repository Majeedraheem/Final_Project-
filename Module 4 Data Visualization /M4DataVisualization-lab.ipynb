### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
conn = sqlite3.connect("m4_survey_data.sqlite") # open a database connection
```

```python
df_ = pd.read_sql_query("SELECT * FROM master", conn)
```

```python

# Filter data for ages between 30 and 35
df_filtered_age = df_[(df_['Age'] >= 30) & (df_['Age'] <= 35)][['WorkWeekHrs', 'CodeRevHrs', 'Age']]

# Compute median values for each age group
median_df = df_filtered_age.groupby('Age').median().reset_index()

# Plotting
plt.figure(figsize=(10, 6))

# Create a stacked bar chart
median_df.set_index('Age').plot(kind='bar', stacked=True, color=['skyblue', 'lightgreen'], edgecolor='black')

# Add labels and title
plt.xlabel('Age')
plt.ylabel('Median Hours')
plt.title('Median Work Hours for Ages 30 to 35')
plt.legend(title='Hours Type', loc='upper left', bbox_to_anchor=(1, 1))  # Position legend outside the plot

# Display the plot
plt.tight_layout()  # Adjust layout to fit labels
plt.savefig('median_work_hours_30_35.png',)  # Save the figure
plt.show()
# Close the figure to free up memory
plt.close()
```

```python
Q1 = df_['Age'].quantile(0.25)
Q3 = df_['Age'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(f'Lower Bound: {lower_bound}')
print(f'Upper Bound: {upper_bound}')

# Find outliers
outliers = df_[(df_['Age'] < lower_bound) | (df_['Age'] > upper_bound)]
print(outliers)
```
